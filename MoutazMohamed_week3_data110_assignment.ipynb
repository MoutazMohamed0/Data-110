```python
import matplotlib.pyplot as plt
print(plt.style.available)
```

```python
import numpy as np
import pandas as pd
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
x = np.linspace(-2, 4, 100)
plt.plot(x, x**2 - 2*x + 1)
```

```python
plt.plot(x, x**3, color='blue', label='X^3')
plt.plot(x, np.exp(x), color='yellow', label='e^x')
plt.plot(x, np.log(x, where=(x>0)), color='red', label='log(x)')
plt.xlim(0, 5)
plt.ylim(0, 10)

plt.legend()
```

```python
plt.plot(x, x**2, color='r', linestyle='solid')
plt.plot(x, x**3, color='0.75', linestyle='dashed')
plt.plot(x, x**4, color='#FFDD44', linestyle='dashdot')
plt.plot(x, x**5, color='blue', linestyle='dotted')
plt.xlim(0, 1.5)
plt.ylim(0, 5);
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, ((1)/(x**2 + 1)))
plt.xlim(-10, 10)
plt.ylim(-0.5, 1.25)
plt.axis('tight')
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, ((1)/(x**2 + 1)))
plt.xlim(10, -10)
plt.ylim(1.25, -0.5)
plt.axis('equal')
```

```python
plt.plot(x, np.sqrt(x), color='blue', label='sqrt(x)')
x_filtered = x[x >= 0]
plt.plot(x, x**(1/3), color='red', label='x^(1/3)')
plt.title("sqrt(x) & x^(1/3)")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = x*np.sin(x)
plt.plot(x, y, '-ok', color='red',
         markersize=7, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='black',
         markeredgewidth=2)
plt.xlim(0, 13)
plt.ylim(-12, 9)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
np.arange(-2*np.pi, 2*np.pi, np.pi/4)
y = x*np.cos(x)
plt.plot(x, y, '-p', color='blue', markerfacecolor='red')
```

```python
plt.style.use('ggplot')
plt.plot(x, np.tan(x), color='#DC143C', label='tan(x)')
plt.plot(x, np.arctan(x), color='#808000', label='arctan(x)')
plt.xlim(-1.5, 1.5)
plt.ylim(-5, 5)
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
```

I chose the functions tan(x) and arctan(x) because the functions are inverses of each other and it is interesting to see how the lines are related when visualized. The x and y limits that I chose allow you to see that the functions intersect at 0. I chose ggplot because of its ease of readability, its dark background enhances the visual appeal of the graphs and makes the graphs more pleasing to look at when compared to seaborn-whitegrid styles.

```python
plt.style.use('default')
x = np.linspace(-2, 4, 1000)
plt.plot(x, x**2 - 2*x + 1)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
x = np.linspace(-2, 4, 1000)
plt.plot(x, x**2 - 2*x + 1)
```

```python
plt.style.use('ggplot')
x = np.linspace(-2, 4, 1000)
plt.plot(x, x**2 - 2*x + 1)
```

The graph made using the default style has a blue line with a plain background that has no grid lines. The graph made using seaborn-whitegrid has a blue line with a background that has a grid line making the graph easier to read. The graph made using ggplot has a red line and grid lines as well as a darker background that allows for better visibility.
