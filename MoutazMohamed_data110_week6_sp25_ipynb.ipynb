```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from joypy import joyplot
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')
df.head()
```

```python
df.info()
```

```python
df_clean = df.dropna()
```

```python
sns.boxplot(x='species', y='body_mass_g', data=df_clean,color="skyblue")
plt.title = plt.gca().set_title
plt.title('Penguin Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

The Gentoo species of penguin has the largest median body mass.

The Gentoo species also seems to have the widest range in body mass.

The Chinstrap penguin has outliers both above and below the IQR for body mass.


```python
import matplotlib.style
matplotlib.style.available
```

```python
sns.violinplot(x='species', y='flipper_length_mm', data=df_clean,color="red")
plt.title = plt.gca().set_title
plt.title('Distribution of Penguin Flipper Length by Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

The Adelie penguin has the most symmetrical distribution.

The Gentoo penguin has the widest spread of flipper lengths.

The median (white dot) for the Adelie penguin shows an even distribution of flipper length. The median for the Chinstrap penguin shows a slightly uneven distribution of flipper length. The median for the Gentoo penguin shows a more uneven distribution for flipper length.

```python
data=df_clean[['island', 'bill_length_mm']]
fig, axes = joyplot(df_clean,
    by='island',
    column='bill_length_mm',
    figsize=(10, 8),
    colormap=plt.cm.viridis,
    ylabelsize=12,
    xlabelsize=12,
    alpha=0.75
)
plt.title = plt.gca().set_title
plt.title('Ridgeline Plot of Penguin Bill Length by Island')
plt.ylabel('Bill Length (mm)')
plt.show()
```

Torgersen island has the most narrow range of bill lengths.

Biscoe island has the most uniform distribution of bill lengths.

Torgersen island penguins generally have shorter bill lengths. Biscoe island penguins have the most broad range of bill lengths. Dream island  has penguins bill lengths have two peaks one peak concentrated at a smaller bill length and the other peak concentrated at a wider bill length.
